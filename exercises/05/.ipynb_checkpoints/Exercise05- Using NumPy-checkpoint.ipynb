## Part A: Basics

We're just going to walk through some basic operations first.

```python
# I would like you to create an array
# of random numbers using the numpy
# random module.  Please do so below.
# I want it to be 1000 rows by 500 columns.
# For the random numbers, I want them to
# be in the range (0,100]
# Use only numpy.random.random(...) for this
# along with some basic math
array1 = (np.random.random((1000,500)) * -100) + 100
```

```python
# Using slicing and broadcasting, halve the first 100 by 100 entries
array1_first100 = array1[0:100,0:100]/2
array1[0:100,0:100] = array1_first100
```

```python
# Make a copy of the array and renormalize the values so that
# the mean is zero, the largest possible max is just less than 1 
# and the minimum is -1. Check your work by using the 
# methods to find the mean, maximum, and minimum.
array2 = ((array1.mean()-array1)/50).copy()
print(array2.mean())
print(array2.min())
print(array2.max())
```

## Part B: Returning to the PointyGame
This is just a quick exercise to play around with data from the PointyGame using NumPy.

```python
# Loading the CSV data
import urllib
from io import StringIO

# Download the CSV file from the web
game_csv = 'http://www.cse.msu.edu/~ruppmatt/itm891/greedy.csv'
with urllib.request.urlopen(game_csv) as response:
    file_lines = response.read().decode('utf-8')

# StringIO makes the file_lines behave like a file stream
game_arr = np.genfromtxt(StringIO(file_lines[1:]),delimiter=',')

# It's usually good practice to avoid using "magic numbers"
# That is to say numbers without any context about what they mean
# Here you can use these variables for indexing
col_game = 0   # The replicate game number
col_round = 1  # The index of the round
col_loc_x = 2  # The creature's x location
col_loc_y = 3  # The creature's y location
col_score = 4  # The creature's score
```

```python
# Find the maximum score for each game; store those values in a numpy array
# You may use a loop and Boolean operations to assist you
game_arr_1 = np.delete(game_arr, (0), axis=0)
max_score = np.zeros(len(list(set(game_arr_1[:,0]))))
for i,game in enumerate(list(set(game_arr_1[:,0]))):
    max_score[i] = game_arr_1[:,4][game_arr_1[:,0]==game].max()
print(max_score)
```

```python
# Find how many rounds each game was played; store those values in a numpy array
# You may use a loop and Boolean operations to assist you
round_count = np.zeros(len(list(set(game_arr_1[:,0]))))
for i,game in enumerate(list(set(game_arr_1[:,0]))):
    round_count[i] = game_arr_1[:,1][game_arr_1[:,0]==game].shape[0]
print(round_count)
```

## Part C: More multi-dimensional arrays

This is just an exercise to get you thinking about high dimension (greater than 2) data.

Again, we haven't covered plotting yet, so follow the instructions below to perform the plots.

```python
# We're going to start by loading in a data file
# This may take some time
vec_csv = 'http://www.cse.msu.edu/~ruppmatt/itm891/a_vector.csv'
with urllib.request.urlopen(vec_csv) as response:
    file_lines = response.read().decode('utf-8')
vec_arr = np.genfromtxt(StringIO(file_lines),delimiter=',')
```

```python
# Reshape the array to be RxCx3
# Shape should be (1004, 800, 3)
vec_arr_3d = vec_arr.reshape((1004,800,3))
```

```python
# Make an array that is the average of the third dimension
vec_arr_3d_avg = vec_arr_3d.mean(axis = 2)
```

```python
# Plot this array
# Use this command plt.imshow(CHANGE_ME, cmap='gray', vmin=0, vmax=255)
# Where CHANGE_ME is the name of the reshaped and averaged array

plt.imshow(vec_arr_3d_avg, cmap='gray', vmin=0, vmax=255)
# Insert your plotting code above this
plt.show()
```

```python
# Transpose and re-plot this array
# And replot the array using the same command as above

plt.imshow(vec_arr_3d_avg.transpose(), cmap='gray', vmin=0, vmax=255)
#Insert your plotting and transposition code above
plt.show()
```

## Part D:  Just a note on complexity

I just want you to execute the cells below and note how long it takes them to execute.  The first cell uses Python's builtin lists to form an array as a list-of-lists.  We're then going to multiply each element in the list by 2.

The cell after uses numpy's methods to do the same.

Note how much faster (by magnitudes) the numpy solution is to the builtin solution.

```python
%%time
# Build the list using loops and builtins
x_list = []
for r in range(0,10000):
    new_list = []
    for c in range(0,10000):
        new_list.append(1)
    x_list.append(new_list)
    
# Loop through the list of lists and multiply everything by 2
for r in range(0,10000):
    for c in range(0,10000):
        x_list[r][c] *= 2
```

```python
%%time
# Using numpy's methods
x_arr = np.ones((10000,10000))
x_arr *= 2
```
